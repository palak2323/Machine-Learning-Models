#  IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Automobile_data1.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
data =df.replace('?',np.NAN) 
data.isnull().sum()
```

```python
data
```

```python
#Handleing missing values
#fill missing data of price, horsepower, peak-rpm, bore, stroke with the respective column mean
#Fill missing data category Number of doors with the mode of the column i.e. Four
```

```python
temp = df[df['price']!='?']
normalised_mean = temp['price'].astype(int).mean()
df['price'] = df['price'].replace('?',normalised_mean).astype(int)

temp = df[df['horsepower']!='?']
normalised_mean = temp['horsepower'].astype(int).mean()
df['horsepower'] = df['horsepower'].replace('?',normalised_mean).astype(int)

temp = df[df['peak-rpm']!='?']
normalised_mean = temp['peak-rpm'].astype(int).mean()
df['peak-rpm'] = df['peak-rpm'].replace('?',normalised_mean).astype(int)

temp = df[df['bore']!='?']
normalised_mean = temp['bore'].astype(float).mean()
df['bore'] = df['bore'].replace('?',normalised_mean).astype(float)

temp = df[df['stroke']!='?']
normalised_mean = temp['stroke'].astype(float).mean()
df['stroke'] = df['stroke'].replace('?',normalised_mean).astype(float)


df['num-of-doors'] = df['num-of-doors'].replace('?','four')
```

```python
df.head()
```
